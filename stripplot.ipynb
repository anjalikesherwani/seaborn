```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.stripplot(x='time',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='tip',data=df)
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='time',y='total_bill',data=df, hue='sex')
```

```python
sns.stripplot(x='time',y='total_bill',data=df,hue='smoker')
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='time',y='total_bill',data=df,hue='smoker',palette='cool')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',
              palette='cool',order=['Sun','Sat','Thur','Fri'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',
              palette='cool',order=['Sun','Sat','Thur','Fri'],
             orient='v',size=15)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',
              color='yellow',order=['Sun','Sat','Thur','Fri'],
             orient='v',size=15,)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',
              palette='cool',marker='D',order=['Sun','Sat','Thur','Fri'],
             orient='v',size=15)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',
              palette='cool',marker='D',order=['Sun','Sat','Thur','Fri'],
             orient='v',size=10,alpha=0.4)
plt.show()
```

```python
sns.boxplot(y='day',x = 'tip',data=df,hue='sex',color='g',
            order=['Sun','Sat','Thur','Fri'],showmeans=True,
           meanprops={"marker":"d","markeredgecolor":"red"})
```
