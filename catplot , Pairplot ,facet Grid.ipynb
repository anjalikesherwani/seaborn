```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
```

```python
df
```

```python
df.head(1)
```

Facet Grid

```python
figr = sns.FacetGrid(df)
figr.map(plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df)
figr.map(plt.plot, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df , col = 'time')
figr.map(plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df,col='time')
figr.map(plt.bar, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='day')
figr.map(plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df,col='time')
figr.map(plt.scatter, 'smoker','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='day',hue = 'sex')
figr.map(plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='day',hue = 'sex')
figr.map(plt.scatter, 'total_bill','tip')
figr.add_legend()
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex',hue = 'day')
figr.map(plt.scatter, 'total_bill','tip')
figr.add_legend()
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex',hue = 'day')
figr.map(plt.scatter, 'total_bill','tip').add_legend()
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex',hue = 'day',palette = 'cool')
figr.map(plt.scatter, 'total_bill','tip').add_legend()
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex',hue = 'day', palette = 'copper')
figr.map(plt.scatter, 'total_bill','tip').add_legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df,vars=['total_bill','tip'], hue = 'smoker')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df,vars=['size','tip'], hue = 'smoker')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill'],palette='Accent')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent',kind='reg')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent',kind='hist')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent',kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent',kind='kde')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue = 'sex', hue_order=['Female','Male'],
        x_vars=['total_bill','size'],palette='Accent',
             kind='kde',markers = 'D')
plt.show()
```
